# API Distribution

```python
import json
import re
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import tiktoken
tokenizer = tiktoken.encoding_for_model("gpt-4")
```

```python
with open('/home/zinccat/datasets/data/data/toolllama_G123_dfs_eval.json', 'r') as file:
    data = json.load(file)
```

```python
data = pd.DataFrame(data)
data.head()
```

```python
length = []
turns = []
for conv in data['conversations']:
    cnt = 0
    for turn in conv:
        if turn['from'] == 'system':
            pass
        if turn['from'] == 'assistant':
            length.append(len(tokenizer.encode(turn['value'])))
            cnt += 1
    turns.append(cnt)
# print('Average length: ', np.mean(length))
# print('Average turns: ', np.mean(turns))
```

```python
# plot length distribution
plt.hist(length, bins=100)
plt.title('Length distribution')
plt.xlabel('Length')
plt.ylabel('Count')
plt.savefig('length.png')
plt.show()
```

```python
# draw box plot for length group by turns
plt.figure(figsize=(10, 5))
plt.boxplot([length[i] for i in range(len(length)) if turns[i] == 1], positions=[1], widths=0.5)
plt.boxplot([length[i] for i in range(len(length)) if turns[i] == 2], positions=[2], widths=0.5)
plt.boxplot([length[i] for i in range(len(length)) if turns[i] == 3], positions=[3], widths=0.5)
plt.boxplot([length[i] for i in range(len(length)) if turns[i] == 4], positions=[4], widths=0.5)
plt.boxplot([length[i] for i in range(len(length)) if turns[i] == 5], positions=[5], widths=0.5)
plt.boxplot([length[i] for i in range(len(length)) if turns[i] == 6], positions=[6], widths=0.5)
plt.boxplot([length[i] for i in range(len(length)) if turns[i] == 7], positions=[7], widths=0.5)
```

```python
#group by turns and calculate average length
turns = pd.DataFrame(turns)
length = pd.DataFrame(length)
turns['length'] = length
turns = turns.groupby(0).mean()
turns = turns.reset_index()
plt.scatter(turns[0], turns['length'])
plt.xlabel('Turns')
plt.ylabel('Average System prompt length')
plt.savefig('turns.pdf')
plt.show()
```

```python
length = np.array(length)
print('Mean: ', np.mean(length))
print('Median: ', np.median(length))
hist = plt.hist(length, bins=100)
plt.xlabel('Length of assistant response (single round, in tokens)')
plt.ylabel('Number of responses')
plt.title('Distribution of assistant response lengths')
plt.savefig('assistant_response_lengths.pdf')
plt.show()
```

```python
sentences = []
for indexed_data in data:
    sentences.extend([sentence for sentence in indexed_data['conversations'] if sentence["from"] == "assistant"])
```

```python
processed_sentences = []
for sentence in sentences:
    processed_sentences.append(sentence["value"])
```

```python
pattern = r'\nAction: ([^\n]+)'
full_apis = [re.search(pattern, sentence).group(1) for sentence in processed_sentences]
filtered_apis = [string for string in full_apis if string != 'Finish']
apis = [string.split('_for_', 1)[1] if '_for_' in string else string for string in filtered_apis]
apis = [s.replace("_", "-") for s in apis]
```

### Mapping & Generating Dist

```python
len(mapping_dict)
```

```python
unique_api = list(set(apis))
len(unique_api)
```

```python
delays = []

for s in apis:
    for key in mapping_dict:
        if set(s).issubset(set(key)):
            delays.append(mapping_dict[key])
            break

print(delays)
```

```python
cleaned_delays = [num for num in delays if num == num]
```

```python
len(cleaned_delays)
```

```python
cleaned_delays
```

```python
converted_delays = [int(delay.replace(' ms', '').replace(',', '')) for delay in cleaned_delays]
len([i for i in converted_delays if i < 100]) / len(converted_delays)
```

```python
data_sorted = np.sort(converted_delays)

cdf = np.arange(1, len(data_sorted) + 1) / len(data_sorted)

plt.figure(figsize=(8, 6))
plt.plot(data_sorted, cdf, linestyle='-', color='blue')  # Using a solid line for the plot
plt.title('Distribution of Delays From All API Calls from ToolBench')
plt.xlabel('Delay (ms)')
plt.ylabel('CDF')
plt.xlim(0, 2000)
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.hist(converted_delays, bins=len(set(converted_delays)) * 10, edgecolor='black')
plt.xlabel('Delay (ms)')
plt.ylabel('Frequency')
plt.title('Distribution of Delays From All API Calls from ToolBench')
plt.xlim(0, 5000)
plt.show()
```

### All possible API delay distribution

```python
all_apis = list(mapping_dict.values())
```

```python
cleaned_all_apis = [num for num in all_apis if num == num]
len(cleaned_all_apis)
```

```python
converted_delays = [int(delay.replace(' ms', '').replace(',', '')) for delay in cleaned_all_apis]
converted_delays = [converted_delay for converted_delay in converted_delays if converted_delay > 5]

plt.figure(figsize=(10, 6))
plt.hist(converted_delays, bins=len(set(converted_delays)) * 10, edgecolor='black')
plt.xlabel('Delay (ms)')
plt.ylabel('Frequency')
plt.title('All API Distribution of Delays From Rapid API Platform')
plt.xlim(0, 5000)
plt.show()
```
